```python
def phi(n,x):
    print x #debug
    if(math.sin(.5*x) !=0):
        print'no'#debug
        y=math.sin((n+1/2.)*x)/(2.*math.sin(.5*x))
    else :
        print'yes'#debug
        y=0.
    return abs(y)
```

```python
def Simpsons (k) :
    s=math.pi/6.*(4.*phi(k,math.pi*.5)+phi(k,math.pi))
    return s
```
